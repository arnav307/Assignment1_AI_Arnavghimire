```python
# Task 3. Outlier Detection:
```

```python
# • Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.
def detect_outliers(df, column):
  Q1= df[column].quantile(0.25)
  Q3= df[column].quantile(0.75)
  IQR= Q3-Q1
  lower_bound= Q1-1.5*IQR
  upper_bound= Q3+1.5*IQR
  outliers= df[(df[column]<lower_bound) | (df[column]>upper_bound)]
  return outliers

gni_outliers=detect_outliers(southAsia_df,'gross_inc_percap')
hdi_outliers=detect_outliers(southAsia_df,'hdi')

outliers=pd.concat([gni_outliers,hdi_outliers])
outliers[['country','hdi','gross_inc_percap']]
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(
    x='gross_inc_percap',y='hdi',data=southAsia_df,
    color='blue',label='Southasian countries'
)
plt.scatter(
    x='gross_inc_percap',y='hdi',data=outliers,
    color='red',label='Outliers'
)
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outlier Detection: GNI per Capita vs HDI (South Asia)")
plt.legend()
plt.show()
```

```python
# Task 5. HDI Disparity:
# Compute the range (max – min) of HDI for each region.
# Compute the coefficient of variation (CV = std/mean) for HDI.
#  Identify which region exhibits more variation in HDI.
```

```python
# Identify which region exhibits more variation in HDI.
if sa_hdi_cv > me_hdi_cv:
    print("South Asia exhibits greater HDI variation.")
else:
    print("Middle East exhibits greater HDI variation.")
```

```python
# 7. Outlier Detection:
```

```python
# • Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
# the 1.5 × IQR rule.
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# South Asia
sa_hdi_outliers = detect_outliers(southAsia_2020_2022, "hdi")
sa_gni_outliers = detect_outliers(southAsia_2020_2022, "gross_inc_percap")
sa_outliers = pd.concat([sa_hdi_outliers, sa_gni_outliers]).drop_duplicates()

# Middle East
me_hdi_outliers = detect_outliers(middleEast_2020_2022, "hdi")
me_gni_outliers = detect_outliers(middleEast_2020_2022, "gross_inc_percap")
me_outliers = pd.concat([me_hdi_outliers, me_gni_outliers]).drop_duplicates()

print("South Asia outliers:\n", sa_outliers[["country","year","hdi","gross_inc_percap"]])
print("---------------------------------------------------------------------")
print("Middle East outliers:\n", me_outliers[["country","year","hdi","gross_inc_percap"]])


```
